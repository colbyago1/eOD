```python
#import modules need for analysis
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt 
```

```python
csv_file_name = '/Users/colbyagostino/projects/eOD/NNK/combined.csv'

# adjust UNBOUND_CUTOFF based on pdb
UNBOUND_CUTOFF = -690
RATIO_CUTOFF = 1

#load the data into a pandas dataframe
df = pd.read_csv(csv_file_name, names='PDB Unbound Ratio RMSD'.split(), index_col=False).dropna()

#filter values
filtered_df = df.query(f'Unbound<{UNBOUND_CUTOFF} and Ratio>{RATIO_CUTOFF}')
```

```python
col1 = "RMSD" # x
col2 = "Ratio" # y
plt.figure()
plt.scatter(df[col1], df[col2])
plt.title(f"{col2} vs {col1}")
plt.xlabel(col1)
plt.ylabel(col2)
```

```python
col1 = "RMSD" # x
col2 = "Ratio" # y
plt.figure()
plt.scatter(filtered_df[col1], filtered_df[col2])
plt.title(f"{col2} vs {col1}, filtered")
plt.xlabel(col1)
plt.ylabel(col2)
```

```python
col1 = "Mutations" # x
col2 = "Ratio" # y

df[col1] = df['PDB'].str.count('__')
grouped_data = df.groupby(col1)[col2].apply(list)

plt.figure()
plt.boxplot(grouped_data, labels=grouped_data.index)
plt.title(f"{col2} vs {col1}")
plt.xlabel(col1)
plt.ylabel(col2)
```
