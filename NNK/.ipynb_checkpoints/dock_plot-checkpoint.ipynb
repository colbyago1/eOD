```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
input_file = "combined_score.csv"
df = pd.read_csv(input_file)
```

```python
# I_sc = energy of the interactions across the interface (typically in the range of -2 to -10 REU)
# Fnat = fraction of native contacts
# rms = C-alpha RMSD of the ligand
# Irms = RMSD of the interface atoms
```

```python
col1 = "I_sc" # x
col2 = "total_score" # y
plt.figure()
plt.scatter(df[col1], df[col2])
plt.title(f"{col1} vs {col2}")
plt.xlabel(col1)
plt.ylabel(col2)
```

```python
col1 = "rms" # x
col2 = "I_sc" # y
plt.figure()
plt.scatter(df[col1], df[col2])
plt.title(f"{col1} vs {col2}")
plt.xlabel(col1)
plt.ylabel(col2)
```
